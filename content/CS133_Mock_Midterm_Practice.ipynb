## Data: Bike-lane sightings

Dr. Kennington logs things he sees in the bike lane on his ride to campus.
Run the cell below to create the dataset you'll use for the whole exam.


```python
import numpy as np
import pandas as pd

np.random.seed(1337)

base_items = np.array([
    'left glove', 'right glove', 'a glove', 'gloves',
    'water bottle', 'bottle', 'plastic bottle',
    'traffic cone', 'orange cone',
    'banana peel', 'apple core',
    'soda can', 'energy drink can', 'can',
    'receipt', 'paper', 'flyer',
    'broken glass', 'glass shard',
    'mask', 'face mask',
    'mystery puddle', 'small branch'
])

n = 180

p = np.array([
    0.05,0.05,0.03,0.02,
    0.06,0.05,0.03,
    0.04,0.02,
    0.05,0.03,
    0.06,0.03,0.03,
    0.06,0.03,0.02,
    0.03,0.02,
    0.05,0.03,
    0.04,0.03
], dtype=float)
p = p / p.sum()

items = np.random.choice(base_items, size=n, replace=True, p=p)

start = np.datetime64('2026-01-06T08:05')
day_offsets = np.random.choice(np.arange(0, 45), size=n, replace=True)
time_offsets_min = np.random.randint(0, 35, size=n)
timestamps = start + day_offsets.astype('timedelta64[D]') + time_offsets_min.astype('timedelta64[m]')

weather = np.random.choice(np.array(['clear','cloudy','rainy','snowy']), size=n, replace=True,
                           p=np.array([0.45,0.28,0.22,0.05]))

distance_m = np.round(np.clip(np.random.normal(loc=2100, scale=850, size=n), 0, None), 0)

raw = pd.DataFrame({
    "timestamp": pd.to_datetime(timestamps.astype("datetime64[ns]")),
    "item": items,
    "weather": weather,
    "distance_m": distance_m.astype(int)
})

raw.head()
```
