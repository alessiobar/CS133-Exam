### Dataset description

This dataset contains the reviews of Tinder from the Google Play Store, ranging from July 2013 to February 2026.

It has the following columns:
- `userName`	Name of a User
- `userImage`	Profile Image that a user has
- `content`	This represents the comments made by a user
- `score`	Scores/Rating between 1 to 5
- `thumbsUpCount`	Number of Thumbs up received by a person
- `reviewCreatedVersion`	Version number on which the review is created
- `at`	Created At
- `replyContent`	Reply to the comment by the Company
- `repliedAt`	Date and time of the above reply
- `reviewId`	unique identifier

**Note**: This CSV is large. To keep things fast, we will load a smaller subset of it.

The dataset is publicly available at from https://www.kaggle.com/datasets/shivkumarganesh/tinder-google-play-store-review, but I uploaded the subsetted version on Canvas.
